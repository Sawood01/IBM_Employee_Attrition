## Level 1 : Univariate Analysis 

```python
# Seperating out the categorical and continuous variables 

def seperate_data_types(df):
    categorical = []
    continuous = []
    
    for column in df.columns:
        if df[column].nunique() < 12:    # Hyperparameter of 12 is used for the Analysis of the given data 
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical,continuous

categorical,continuous = seperate_data_types(df)
```

```python
from tabulate import tabulate
table = [categorical,continuous]

print(tabulate({"Categorical":categorical,"Continuous":continuous},headers = ['Categorical','Continuous']))
```

## Level 3 - Multivariate Analysis 

```python
# To investigate the relationship between Attrition, BusinessTravel, and Age
fig, ax = plt.subplots(1,2,figsize=(15,7))
sns.countplot(x='BusinessTravel', hue='Attrition', data=df, ax=ax[0])
ax[0].set_xlabel('Business Travel')
ax[0].set_ylabel('Count')
ax[0].set_title('Attrition by Business Travel')
sns.boxplot(x='BusinessTravel', y='Age', hue='Attrition', data=df, ax=ax[1])
ax[1].set_xlabel('Business Travel')
ax[1].set_ylabel('Age')
ax[1].set_title('Age by Business Travel and Attrition')
plt.show()
```

```python
# Which age group of men has larger churn
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 40, 50, 60, 70])
male_df = df[df['Gender'] == 'Male']
male_df.groupby(['Gender', 'AgeGroup', 'Attrition']).size().unstack().plot(kind='bar', stacked=True)
plt.xlabel('Gender, Age Group')
plt.ylabel('Count')
plt.title('Count of Male Employees by Age Group and Attrition')
plt.show()
```

### ***Interpretation : Male Employees those below 40 years  of age has higher churn rate compared to other age groups***

```python
data = df[['JobRole', 'Gender', 'Attrition']]

# Group the data by JobRole, Gender, and Attrition
grouped = data.groupby(['JobRole', 'Gender', 'Attrition']).size().reset_index(name='counts')

# Create a pivot table to reshape the data
pivot = grouped.pivot(index=['JobRole', 'Gender'], columns='Attrition', values='counts').fillna(0)

# Plot a stacked bar chart to visualize the data
pivot.plot(kind='barh', stacked=True, figsize=(10, 6))

# Set the title and axis labels
plt.title('Attrition by Job Role and Gender')
plt.ylabel('Job Role and Gender')
plt.xlabel('Count')
plt.show()
```
